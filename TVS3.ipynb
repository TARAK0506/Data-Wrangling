#  16. SELECT planes.* , airlines.*  FROM (SELECT DISTINCT carrier,tailnumFROM flights) AS cartail INNER JOIN planes ON cartail.tailnum=planes.tailnum INNER JOIN airlines ON cartail.carrier=airlines.carrier

**two joins need to be done 
two primary keys are 1.tailnum and 2.carrier**

**select distinct pairs 'carrier' and 'tailnum' columns and store them in cartail**

```python
flights.shape
```

```python
cartail = flights[['carrier', 'tailnum']].drop_duplicates()
```

**compute inner join between cartail and planes**

```python
inner_join_cartail_planes= pd.merge(cartail, planes, on='tailnum', how='inner')
```

**now compute innerjoin between above result and airlines**

```python
result_16= pd.merge(inner_join_cartail_planes, airlines, on='carrier', how='inner')
result_16
```

```python
result_16.shape
```
